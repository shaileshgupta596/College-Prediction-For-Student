```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url="https://www.yellowpages.com/search?search_terms=coffee&geo_location_terms=Los+Angeles%2C+CA"
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content,"lxml")
```

```python
g_data = soup.find_all("div",{"class":"info"})
```
